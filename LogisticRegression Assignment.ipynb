```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, roc_curve, roc_auc_score

import warnings
warnings.filterwarnings("ignore")
```

```python
bank = pd.read_csv("bank-full.csv")
```

```python
bank
```

```python
bank.info()
```

```python
data_dum = pd.get_dummies(bank,columns = ["job", "marital", "education", "contact", "poutcome"])
data_dum
```

```python
pd.set_option("display.max.columns", None)
data_dum
```

```python
data_dum.info()
```

```python
data_dum["default"] = np.where(data_dum["default"].str.contains("yes"), 1, 0)
data_dum["housing"] = np.where(data_dum["housing"].str.contains("yes"), 1, 0)
data_dum["loan"] = np.where(data_dum["loan"].str.contains("yes"), 1, 0)
data_dum["y"] = np.where(data_dum["y"].str.contains("yes"), 1, 0)
data_dum
```

```python
data_dum["month"].value_counts()
```

```python
order={'month':{'jan':1,'feb':2,'mar':3,'apr':4,'may':5,'jun':6,'jul':7,'aug':8,'sep':9,'oct':10,'nov':11,'dec':12}}
```

```python
data_dum = data_dum.replace(order)
data_dum
```

```python
data_dum.info()
```

```python
X = pd.concat([data_dum.iloc[:,0:11], data_dum.iloc[:,12:]], axis = 1)
Y = data_dum.iloc[:,11]
```

```python
model = LogisticRegression()
model.fit(X, Y)
```

```python
y_pred = model.predict(X)
y_pred
```

```python
y_pred_df = pd.DataFrame({"actual_y":Y, "y_pred_prob":y_pred})
y_pred_df
```

```python
confusion_matrix = confusion_matrix(Y, y_pred)
confusion_matrix
```

```python
model.predict_proba(X)[:,1]
```

```python
fpr, tpr, thresholds = roc_curve(Y, model.predict_proba(X)[:,1])
plt.plot(fpr, tpr, color = "green")
auc = roc_auc_score(Y, y_pred)

plt.plot(fpr, tpr, color = "green", label = "logit model(area = %0.2f)"%auc)
plt.plot([0,1], [0,1], "k--")
plt.xlabel("False Positive Rate or [1 - True Negative Rate]")
plt.ylabel("True Positive Rate")
plt.show()

print("auc accuracy:", auc)
```
